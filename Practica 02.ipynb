```python
# === PASO 1: Importar librerías ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === PASO 2: Leer la base de datos desde el enlace ===
url = 'https://drive.google.com/uc?id=1BE-dfkrb6oyLKDuqXAq2fDYMkDz2f9hM&export=download'
df = pd.read_csv(url)
```

```python
# === PASO 3: Vista previa de los datos ===
print("Primeras 5 filas del dataset:")
df.head()
```

```python
#==Información general del dataset

print("Resumen general:")
df.info()

print("\nResumen estadístico:")
df.describe(include='all')
```

```python
#==Verificar valores faltantes

print("Valores nulos por columna:")
df.isnull().sum()
```

```python
#==Análisis exploratorio de ejemplo
#==a. ¿Cuántos productos por categoría?

df['Category'].value_counts().head(10).plot(kind='bar', figsize=(10, 5), title='Top 10 categorías de productos')
plt.xlabel("Categoría")
plt.ylabel("Cantidad")
plt.show()
```

```python
#==b. Precio promedio por marca

df.groupby('Brand')['Price'].mean().sort_values(ascending=False).head(10).plot(kind='barh', title='Precio promedio por marca')
plt.xlabel("Precio promedio")
plt.ylabel("Marca")
plt.show()
```

```python
#==c. ¿Cuántos productos están disponibles?

df['Availability'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(5, 5), title='Disponibilidad de productos')
plt.ylabel('')
plt.show()
```
